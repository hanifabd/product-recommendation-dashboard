```python
import csv
from datetime import datetime, timedelta
import random

# Function to generate a random date within a given range
def random_date(start_date, end_date):
    return start_date + timedelta(days=random.randint(0, (end_date - start_date).days))

# Set the start and end dates for purchase_date
start_date = datetime(2023, 1, 1)
end_date = datetime(2023, 1, 31)

# Generate 100 rows of data
rows = []
for _ in range(300):
    customer_id = random.randint(1, 50)
    product_id = random.randint(101, 130)
    purchase_date = random_date(start_date, end_date).strftime('%Y-%m-%d')
    rows.append([customer_id, product_id, purchase_date])

# Append the generated rows to the existing data
existing_data = [
    [1, 101, '2023-01-01'],
    [1, 105, '2023-01-05'],
    [2, 102, '2023-01-02'],
    [3, 103, '2023-01-03'],
    [4, 104, '2023-01-04'],
    [5, 101, '2023-01-05']
]

all_data = existing_data + rows

# Write the data to a new CSV file
with open(root+'generated_purchase_history.csv', 'w', newline='') as csvfile:
    csvwriter = csv.writer(csvfile, delimiter=';')
    csvwriter.writerow(['customer_id', 'product_id', 'purchase_date'])
    csvwriter.writerows(all_data)
```

```python
import csv
import random

# Generate 25 rows of data with unique product IDs from 101 to 130
additional_rows = []
for product_id in range(106, 131):
    category_options = ['Electronics', 'Clothing', 'Home & Kitchen', 'Beauty']
    category = random.choice(category_options)
    price = random.uniform(20, 500)  # Adjust the price range as needed
    ratings = round(random.uniform(3.0, 5.0), 1)
    additional_rows.append([product_id, category, price, ratings])

# Append the generated rows to the existing data
existing_data = [
    [101, 'Electronics', 500, 4.5],
    [102, 'Clothing', 50, 3.8],
    [103, 'Home & Kitchen', 200, 4.2],
    [104, 'Beauty', 30, 4.0],
    [105, 'Electronics', 800, 4.8]
]

all_data = existing_data + additional_rows

# Write the data to a new CSV file
with open(root+'generated_product_details.csv', 'w', newline='') as csvfile:
    csvwriter = csv.writer(csvfile, delimiter=';')
    csvwriter.writerow(['product_id', 'category', 'price', 'ratings'])
    csvwriter.writerows(all_data)
```
